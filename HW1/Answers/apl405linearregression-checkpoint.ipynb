```python
import numpy as np
import matplotlib.pyplot as plt

# loading the Mat file using scipy
data = scipy.io.loadmat('house_data.mat')

# features and target from training data
features_train = data['Xtrain']
target_train = data['Ytrain'].flatten()

# Add bias term to the data
X_train = np.hstack((features_train, np.ones((features_train.shape[0], 1))))  # Adding a column of ones for bias

# Compute the weight vector using the closed-form solution
XTX_inv = np.linalg.inv(X_train.T @ X_train)
weights = XTX_inv @ X_train.T @ target_train

# Make predictions on the training set
predictions_train = X_train @ weights

# Compute mean squared error on the training set
mse_train = np.mean((predictions_train - target_train) ** 2)
print("Mean Squared Error on Training Set:", mse_train)

# features and target from test data
features_test = data['Xtest']
target_test = data['Ytest'].flatten()

# Add bias term to test features
X_test = np.hstack((features_test, np.ones((features_test.shape[0], 1))))

# Make predictions on the validation set
predictions_test = X_test @ weights

# Compute mean squared error on the test set
mse_test = np.mean((predictions_test - target_test) ** 2)
print("Mean Squared Error on Test Set:", mse_test)

# Plot predicted vs. actual values
plt.scatter(target_test, predictions_test, alpha=0.5)
plt.plot([min(target_test), max(target_test)], [min(target_test), max(target_test)], color='red', linestyle='--')
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.title("Predicted vs. Actual Values")
plt.show()
```

```python

plt.plot(weights[:-1][::])  
plt.ylabel('Parameter Index') 
plt.xlabel('Parameter Value') 
plt.title('Model Coefficients Visualization')  
plt.show()
```

```python

residuals = predictions_test - target_test

plt.hist(residuals, bins=10, edgecolor='black',alpha=0.7)
plt.xlabel('Residuals')
plt.ylabel('Frequency')
plt.title('Histogram of Residuals')
plt.show()
```
